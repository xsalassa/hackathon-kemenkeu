```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Hackaton Processed 2.csv")
```

```python
df.info()
```

```python
df_nik_unique = df.drop_duplicates(subset = ['nik_hashed', 'checkin_timestamp', 
                                         'outlet_sub_category', 'city_dagri_code'], 
                               keep = 'last').reset_index(drop = True)
```

```python
df_nik_unique.info()
```

```python
len(df_nik_unique["nik_hashed"].unique())
```

```python
df_nik_unique = df_nik_unique.sort_values(by='checkin_timestamp')
```

```python
df_nik_unique
```

```python
user_color_status = df_nik_unique.groupby("nik_hashed")["user_color_status"].unique()
```

```python
user_color_status_dict = []
nik_with_color_status = []
nik_with_color_red = []
nik_with_color_green = []
nik_with_color_black = []
nik_with_color_yellow = []
for i, j in user_color_status.items():
    
    if len(j) > 1:
        nik_with_color_status.append(i)
        dix = {}
        dix['nik'] = i
        for x, k in enumerate(j):
            dix[x] = k
        user_color_status_dict.append(dix)
    else:
        if j[0] == "RED":
            nik_with_color_red.append(i)
        elif j[0] == "YELLOW":
            nik_with_color_yellow.append(i)
        elif j[0] == "BLACK":
            nik_with_color_black.append(i)
        elif j[0] == "GREEN":
            nik_with_color_green.append(i)
```
